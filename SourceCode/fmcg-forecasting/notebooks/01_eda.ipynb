```python
# Generate synthetic daily_timeseries.csv using real SKU and location IDs

import numpy as np

daily_path = DATA_ROOT / "daily_timeseries.csv"

if daily_path.exists():
    print("daily_timeseries.csv already exists. Skipping generation.")
else:
    # Use real IDs from master tables
    sku_ids = sku["sku_id"].tolist()
    location_ids = location["location_id"].tolist()

    # Date range (3 months of daily data)
    dates = pd.date_range("2024-01-01", "2024-03-31", freq="D")
    festival_dates = set(festival["date"])

    rows = []
    np.random.seed(42)

    for sku_id in sku_ids:
        base_demand = np.random.randint(20, 80)
        for loc_id in location_ids:
            loc_factor = np.random.uniform(0.8, 1.2)
            for i, d in enumerate(dates):
                # Weekly seasonality
                weekly = 10 * np.sin(2 * np.pi * i / 7)
                # Festival boost
                fest_boost = 15 if d in festival_dates else 0
                # Random noise
                noise = np.random.normal(0, 5)
                # Final units
                units = max(0, int(base_demand * loc_factor + weekly + fest_boost + noise))

                rows.append({
                    "date": d,
                    "sku_id": sku_id,
                    "location_id": loc_id,
                    "units_sold": units,
                    "price": round(np.random.uniform(8, 15), 2),
                    "promo_flag": int(np.random.random() < 0.15),
                })

    df_daily = pd.DataFrame(rows)
    df_daily.to_csv(daily_path, index=False)

    print("Created daily_timeseries.csv at:", daily_path)
    print("Shape:", df_daily.shape)
    print("Date range:", df_daily["date"].min(), "→", df_daily["date"].max())
    print("Columns:", list(df_daily.columns))
    df_daily.head()
```
